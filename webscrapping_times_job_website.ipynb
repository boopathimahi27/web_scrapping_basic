```python
#importing the libraries

from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
```

```python
#to get the request link
req_link = requests.get('https://www.timesjobs.com/candidate/job-search.html?searchType=personalizedSearch&from=submit&txtKeywords=python&txtLocation=').content
soup=bs(req_link,'lxml')
```

```python
#creating the necessary variables to fetch the required variables
skill_set=[]
exp_year=[]
date=[]
soup.prettify()
c_name=[]
Location=[]
```

```python
#using the beautiful soup to get the informations
names = soup.find_all('li',class_='clearfix job-bx wht-shd-bx')
```

```python
names
```

```python
#fetching each values
for x in names:
    company_name=x.find('h3','joblist-comp-name')
    c_name.append(company_name.text.strip())
    aaa=x.find("li").text
    years=aaa.lstrip("card_travels")
    exp_year.append(int(years[0]))
    aab=x.find("span").text
    if aab =="":
        Location.append("NaN")
    else:
        Location.append(aab)
    aac=x.find('span',class_="srp-skills").text.replace("  ","").strip()
    skill=list(aac.split(','))
    skill_set.append(skill)
    aad=x.find('div','list-action')
    posted_date=aad.find('span','sim-posted').text.strip('\n')
    date.append(posted_date)
```

```python
#fetched ones to be stored in key values format
value={'Company_Name':c_name,"Min_year_exp":exp_year,"skill":skill_set,"location":Location,"Date":date}
```

```python
#converting it into a dataset
Dataset=pd.DataFrame(value)
```
